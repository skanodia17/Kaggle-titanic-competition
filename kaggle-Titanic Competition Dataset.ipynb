#Naive bayes Algorithm

#Decision Tree Algorithm

```python

d_tree=tree.DecisionTreeClassifier(criterion = 'gini', max_depth = 3, max_leaf_nodes = 6,
                                    min_samples_leaf = 10, min_samples_split = 2)
```

```python
d_tree.fit(X_train,y_train)
```

```python
predictions3=d_tree.predict(X_test)
```

```python
print(classification_report(y_test,predictions3))
```

#Random Forest Algorithm

```python

r_forcls = RandomForestClassifier(n_estimators = 10, n_jobs = 1, random_state = 50)
r_forcls.fit(X_train,y_train)
```

```python
predictions4 = r_forcls.predict(X_test)
```

```python
print(classification_report(y_test,predictions4))
```

#Support Vector machine

```python
svc = SVC()
svc.fit(X_train,y_train)
predictions5 = svc.predict(X_test)
print(classification_report(y_test,predictions5))
```
